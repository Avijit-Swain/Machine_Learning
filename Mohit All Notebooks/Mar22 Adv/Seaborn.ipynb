```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
iris['petal_length']
```

```python
sns.histplot(data=iris, x = 'petal_length', bins=20)
plt.show()
```

```python
sns.histplot(iris['petal_length'].values, bins=20)
plt.show()
```

```python
sns.kdeplot(data=iris, x='petal_length', color='red')
plt.show()
```

```python
iris.head()
```

```python
sns.scatterplot(data=iris, x='petal_length', y ='petal_width')
plt.show()
```

```python
sns.scatterplot(data=iris, x='petal_length', y ='petal_width', hue = 'species')
plt.show()
```

```python
iris['species'].value_counts()
```

```python
sns.histplot(data=iris, x = 'petal_length', bins=20, hue='species', stat='density')
plt.show()
```

```python
sns.kdeplot(data=iris, x='petal_length', hue='species')
plt.show()
```

## JointPlot

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='reg')
plt.show()
```

### Hexabin Plotting

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', kind='hex')
plt.show()
```

## Boxplot

```python
sns.boxplot(data=iris, y='sepal_width')
plt.show()
```

```python
sns.violinplot(data=iris, x='species', y = 'sepal_width')
plt.show()
```

## Correlations 

```python
corr = iris.corr()
corr
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
sns.jointplot(data=iris, x='petal_length', y='petal_width', kind='hex', gridsize=15)
plt.show()
```

```python
sns.heatmap(corr, cmap="coolwarm", annot=True, )
plt.show()
```
