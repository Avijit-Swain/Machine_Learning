```python
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (18,10)
from matplotlib.pyplot import cm
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.keys()
```

```python
X = digits.data
Y = digits.target
```

```python
X.shape
```

```python
digits.images.shape
```

```python
plt.gray()
plt.imshow(digits.images[1000], cmap=plt.cm.gray_r)
```

```python
from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.husl_palette(len(np.unique(Y))))
```

##PCA

```python
from sklearn.decomposition import PCA

z1 = PCA(n_components=2).fit_transform(X)
```

```python
z1.shape
```

```python
fig, ax = plt.subplots()
im = ax.scatter(z1[:,0], z1[:,1], s=50, c=Y, cmap=cmap, edgecolor='none')
cbar = fig.colorbar(im, ax=ax,label='Digit')
```

##t-SNE

```python
from sklearn.manifold import TSNE

z2 = TSNE(n_components=2, random_state=0).fit_transform(X)
```

```python
z2.shape
```

```python
fig, ax = plt.subplots()
im = ax.scatter(z2[:,0], z2[:,1], s=25, c=Y, cmap=cmap, edgecolor='none')
cbar = fig.colorbar(im, ax=ax,label='Digit')
```
