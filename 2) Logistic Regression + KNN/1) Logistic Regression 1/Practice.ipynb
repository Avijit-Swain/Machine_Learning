```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def hypothesis(X,weights):
    return sigmoid(np.dot(X,weights))
```

```python
def error(X,y,weights):
    model = hypothesis(X, weights)
    err = np.mean((y*np.log(model) + (1-y)*np.log(1 - model)))
    
    return -err
```

```python
def gradient(X,y,weights):
    model = hypothesis(X, weights)
    grad = (np.dot(X.T, (model - y)))
    return grad/X.shape[0]
```

```python
def gradient_descent(X,y,max_iter=200,lr=0.5):
    weight = np.zeros((X.shape[1], 1))
    error_list = []
    
    for _ in range(max_iter):
        e = error(X, y, weight)
        error_list.append(e)
        
        grad = gradient(X, y, weight)
        
        #Update Rule
        weight = weight - lr*grad
    
    return (weight, error_list)
```
